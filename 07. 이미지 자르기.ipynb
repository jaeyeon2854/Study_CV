# 이미지 자르기 (동영상도 같은 방식으로 진행)

## 영역을 잘라서 새로운 윈도우(창)에 표시

```python
import cv2
img = cv2.imread('img.jpg')
#img.shape (391, 640, 3)

# 이미지 자르기 : crop = img[세로범위, 가로범위]
crop = img[100:200, 200:400]

cv2.imshow('img', img) # 원본 이미지
cv2.imshow('crop' , crop) # 잘린 이미지
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### 영역을 잘라서 기존 윈도우에 표시

```python
import cv2
img = cv2.imread('img.jpg')

crop = img[100:200, 200:400]

img[100:200,400:600]=crop 

cv2.imshow('img', img) # 원본 이미지
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
